```python
import tensorflow as tf
import cv2

CATEGORIES = ["Cat", "Dog"]
```

```python
model = tf.keras.models.load_model("models/64x3-CNN.model")
```

```python
IMG_SIZE = 50
def test_image_preparation(img_path):
    img_array = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
    new_array = cv2.resize(img_array, (IMG_SIZE, IMG_SIZE))
    return new_array.reshape(-1, IMG_SIZE, IMG_SIZE, 1)
```

```python
cat_img_path = './PetImages/Cat/1234.jpg'
dog_img_path = './PetImages/Dog/1234.jpg'

import matplotlib.pyplot as plt

img_arr = img_array = cv2.imread(cat_img_path, cv2.IMREAD_GRAYSCALE)
plt.imshow(img_arr, cmap='gray')
plt.show()
```

```python
prediction = model.predict([test_image_preparation(cat_img_path)])
```
